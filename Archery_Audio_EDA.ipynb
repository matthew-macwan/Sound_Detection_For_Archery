```python
### CREATE WAV FILES FROM MP4 or MOV files
```

```python
# create a folder in the current directory where we will have the new wav files stored 

main_path = '/Users/matthew_macwan/Downloads/CIS/Archery_Project'
  
path = os.path.join(main_path, 'audio_clips')
os.mkdir(path)
```

```python
def MP4_to_WAV(inpath, outpath):
    
    i = 1
    
    for video_path in os.listdir(inpath):

        full_input_path = os.path.join(inpath, video_path)
        
        clip = AudioSegment.from_file(full_input_path, format="mp4")
        clip.export(outpath+"/archery_"+str(i)+'.wav', format="wav")
        
        print('MP4.video_',i, '--- Export successful\n')
        
        i = i+1
        
        
```

```python
inpath = '/Users/matthew_macwan/Downloads/CIS/Archery_Project/video_clips'
outpath = '/Users/matthew_macwan/Downloads/CIS/Archery_Project/audio_clips'
```

```python
MP4_to_WAV(inpath, outpath)
```
